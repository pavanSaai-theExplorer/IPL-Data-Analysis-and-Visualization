# Data Preparing and Cleaning

```python
# Importing required Libraries

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Reading data using Pandas

ipl_data = pd.read_csv("IPL_Matches(2008-2020).csv")
```

```python
# Getting the datatype for each column

ipl_data.info()
```

```python
# Getting the number of rows and columns

ipl_data.shape
```

```python
# Overview of the data

ipl_data.describe()
```

```python
# Displaying the top 5 records

ipl_data.head()
```

```python
# Removing unimportant fields 

ipl_data.drop(columns = ["umpire1","umpire2","method"], inplace = True)
```

```python
# Displaying the processed data

ipl_data
```

```python
# Displaying the columns

print(ipl_data.columns)
```

```python
# Displaying all the teams

print(ipl_data.team1.unique())   ## print(ipl_data.team2.unique())
```

```python
# Displaying all the cities

print(ipl_data.city.unique())
```

```python
# Displaying all the venues

print(ipl_data.venue.unique())
```

```python
# Replacing the old names of the teams with recent names

ipl_data.team1.replace({'Rising Pune Supergiants' : 'Rising Pune Supergiant', 'Delhi Daredevils':'Delhi Capitals','Pune Warriors' : 'Rising Pune Supergiant'},inplace=True)
ipl_data.team2.replace({'Rising Pune Supergiants' : 'Rising Pune Supergiant', 'Delhi Daredevils':'Delhi Capitals','Pune Warriors' : 'Rising Pune Supergiant'},inplace=True)
ipl_data.toss_winner.replace({'Rising Pune Supergiants' : 'Rising Pune Supergiant', 'Delhi Daredevils':'Delhi Capitals','Pune Warriors' : 'Rising Pune Supergiant'},inplace=True)
ipl_data.winner.replace({'Rising Pune Supergiants' : 'Rising Pune Supergiant', 'Delhi Daredevils':'Delhi Capitals','Pune Warriors' : 'Rising Pune Supergiant'},inplace=True)
ipl_data.city.replace({'Bangalore' : 'Bengaluru'},inplace=True)
```

# Exploratory Analysis and Visualization

```python
# Finding the total number of matches played from 2008-2020

ipl_data.id.count()
```

```python
# Finding the number of regualr matches

regular_matches = ipl_data[ipl_data.result != "tie"].count()
regular_matches.result
```

```python
# Finding the number of tie matches

tie_matches = ipl_data[ipl_data.result == "tie"].count()
tie_matches.result
```

```python
ipl_data.city.unique()
```

```python
# Counting the number of matches played in each city

cities_cnt = ipl_data.groupby("city")[["id"]].count()
```

```python
cities_cnt
```

```python
# Arranging in organised manner

plt.figaspect
cities_cnt.rename(columns={"id":"Matches"},inplace=True)
cities_cnt = cities_cnt.sort_values('Matches',ascending=False).reset_index()
cities_cnt
```

```python
# Plotting the Bar Graph

plt.figure(figsize = (15,10))
plt.grid()
plt.title("Number Of Matches Played in each City", fontweight = "bold", fontsize = 25, color = "r")
plt.xlabel("Matches", fontweight = "bold", fontsize = 20, color = "b")
plt.ylabel("City",fontweight = "bold" , fontsize = 20, color = "b")
sns.barplot(x="Matches",y="city",data=cities_cnt);
```

```python
ipl_data.winner.unique()
```

```python
# Getting the total matches won by each team

winner_data = ipl_data.groupby('winner')[['id']].count()
winner_data = winner_data.sort_values('id', ascending=False).reset_index()

winner_data.rename(columns = {'id':'Wins','winner':'Teams'},inplace=True)
winner_data
```

```python
#Plotting bar Graph

plt.figure(figsize = (15,10))
plt.xlabel("Teams", fontweight = "bold", fontsize = 20 )
plt.ylabel("Wins",fontweight = "bold" , fontsize = 20, )
plt.title("Matches Won By each Team",fontweight = "bold", fontsize = 25, color = "r");
plt.bar(winner_data.Teams,winner_data.Wins, color = ['blue','#FFD801','#461B7E','#C11B17','#F660AB','#000080','#F535AA','#F87217','#BCC6CC','#2C04A2','#E04F16','#632B72']);
plt.xticks(rotation = 90);
```

# Who has been awarded with Player of the Match maximum number of times ?

```python
# Finding the number of players has been awarded the Player of Match title

len(ipl_data.player_of_match.unique())
```

```python
player_data = ipl_data.groupby("player_of_match")[["id"]].count()
player_data = player_data.sort_values("id", ascending = False).reset_index()
player_data
```

```python
# Retreiving the top 10 players

player_data.head(10)
```

```python
# Organising the data in proper order

players_data = player_data.head(10).copy()
players_data.rename(columns={"id":"Total_Awards","player_of_match":"Man_Of_the_Match"},inplace=True)
players_data
```

```python
# Plotting the Bar Graph

plt.figure(figsize=(15,10))
plt.title("Top 10 Players with Highest Man Of the Match Titles",fontweight='bold' )
plt.xticks(rotation=90)
# plt.yticks(ticks=np.arange(0,25,5))
plt.ylabel('No. of Awards')
plt.xlabel('Players')
sns.barplot(x=players_data.Man_Of_the_Match,y=players_data.Total_Awards, alpha=0.8);
```

# What was the most preferred decision on winning the Toss (Bat / Field) ?

```python
# Organising the data in proper order

decision_data = ipl_data.groupby('toss_decision')[['id']].count()
decision_data = decision_data.sort_values('id').reset_index()
decision_data.rename(columns={'id':'Total','toss_decision':'Decision'},inplace=True)
decision_data
```

```python
# Plotting the Bar Graph

plt.figure(figsize=(10,5))
plt.title("Preferred Decision",fontweight='bold',fontsize=30)
plt.xlabel('Decision',fontweight='bold',fontsize=30)
plt.ylabel('Total',fontweight='bold',fontsize=30)
plt.grid()
plt.bar(decision_data.Decision, decision_data.Total, color=['#4863A0','#566D7E']);
```
